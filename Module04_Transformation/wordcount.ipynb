```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("test").getOrCreate()
```

```python
spark
```

```python
input_path = 'input.txt'

rdd = spark.sparkContext.textFile(input_path)
out_ = rdd.flatMap(lambda X: X.split(' '))
```

```python
out_.collect()
```

```python
out_ = out_.map(lambda X: (X, 1)).reduceByKey(lambda a,b: a + b)
```

```python
out_.take(20)
```

```python
out_.count()
```
